```python
import pandas as pd;
import matplotlib.pyplot as plt
import numpy as np;
import ssl
lefthanded_data = pd.read_csv("data.csv")
lefthanded_data.head()
```

```python
plt.scatter(lefthanded_data["Age"], lefthanded_data["Male"], label="Male")
plt.scatter(lefthanded_data["Age"], lefthanded_data["Female"], label="Female")

plt.title("Left-Handedness vs Age")
plt.xlabel("Age")
plt.ylabel("Percentage")
plt.legend()
plt.show()
```

```python
lefthanded_data["Birth_year"] = 1986 - lefthanded_data["Age"]
```

```python
lefthanded_data["Mean_lh"] = (lefthanded_data["Male"] + lefthanded_data["Female"])/ 2
```

```python
lefthanded_data.plot( x="Birth_year",
    y="Mean_lh")
plt.title("Mean Left-Handedness vs Birth Year")
plt.xlabel("Birth Year")
plt.ylabel("Mean Left-Handedness (%)")
plt.show()
```

```python
late_1900s_rate = lefthanded_data["Mean_lh"].head(10).mean()
early_1900s_rate = lefthanded_data["Mean_lh"].tail(10).mean()

def P_LH_given_A(ages_of_death, study_year=1986):
    
    ages_of_death = np.array(ages_of_death)
    
    birth_years = study_year - ages_of_death
    
    P_return = np.zeros(len(ages_of_death))
    
    late_mask = birth_years >= 1930
    early_mask = birth_years < 1930
    
    P_return[late_mask] = late_1900s_rate / 100
    P_return[early_mask] = early_1900s_rate / 100
    
    return P_return
```

```python
P_LH_given_A([20, 50, 80])
```

```python
data_url_2 = "https://gist.githubusercontent.com/mbonsma/2f4076aab6820ca1807f4e29f75f18ec/raw/62f3ec07514c7e31f5979beeca86f19991540796/cdc_vs00199_table310.tsv"
```

```python
death_distribution_data = pd.read_csv(
    data_url_2,
    sep='\t',
    skiprows=[1]
)
death_distribution_data.head()
```

```python
death_distribution_data = death_distribution_data.dropna(subset=["Both Sexes"])
```

```python
death_distribution_data.plot(
    x="Age",
    y="Both Sexes",
    kind="line",
    title="Number of Deaths by Age",
    xlabel="Age",
    ylabel="Number of Deaths"
)
```

```python
def P_lh(death_distribution_data, study_year=1990):
    
    P_lh_A = death_distribution_data["Age"].apply(
        lambda age: P_lh_given_A(age, study_year)
    )
    
    p_list = death_distribution_data["Both Sexes"] * P_lh_A
    
    p = p_list.sum()
    
    total = death_distribution_data["Both Sexes"].sum()
    
    return p / total
```

```python
def P_A_given_lh(age, death_distribution_data, study_year=1990):
    
    death_distribution_data.columns = death_distribution_data.columns.str.strip()
    deaths_col = death_distribution_data.columns[1]

    total_deaths = death_distribution_data[deaths_col].sum()

    deaths_at_age = death_distribution_data.loc[
        death_distribution_data["Age"] == age, deaths_col
    ]

    if len(deaths_at_age) == 0:
        return 0

    deaths_at_age = deaths_at_age.values[0]

    P_A = deaths_at_age / total_deaths
    P_LH = P_lh(death_distribution_data, study_year)
    P_LH_given_A = P_lh_given_A(age, study_year)

    return (P_LH_given_A * P_A) / P_LH
```

```python
print(P_lh(death_distribution_data))
print(P_A_given_lh(50, death_distribution_data))
```
